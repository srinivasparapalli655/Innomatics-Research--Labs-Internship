```python
# Question 2:
# Given an integer n, perform the following actions:
# - If n is odd, print "Weird"
# - If n is even and in the inclusive range 2 to 5, print "Not Weird"
# - If n is even and in the inclusive range 6 to 20, print "Weird"
# - If n is even and greater than 20, print "Not Weird"
# Input: A single positive integer n
# Output: "Weird" or "Not Weird" based on the above rules.

# Answer:
n = int(input().strip())

if n % 2 == 1:
    print("Weird")
elif 2 <= n <= 5:
    print("Not Weird")
elif 6 <= n <= 20:
    print("Weird")
else:
    print("Not Weird")
```

```python
# Question 3:
# The program reads two integers a and b from input.
# Print three lines:
# 1) The sum a + b
# 2) The difference a - b
# 3) The product a * b

# Answer:
a = int(input().strip())
b = int(input().strip())

print(a + b)
print(a - b)
print(a * b)
```

```python
# Question 4 :
# The program reads two integers a and b from input.
# Print two lines:
# 1) The result of integer division a // b
# 2) The result of float division a / b
# No rounding or special formatting is required.

# Answer:
a = int(input().strip())
b = int(input().strip())

print(a // b)
print(a / b)
```

```python
# Question 5: 
# The program reads an integer n from input.
# For all non-negative integers i such that i < n, print i^2 on a separate line.
# Input: A single integer n
# Output: n lines where the i-th line (starting from 0) contains i^2 for i in [0, n-1].

# Answer:
n = int(input().strip())

for i in range(n):
    print(i ** 2)
```

```python
# Question 6:
# Write a function is_leap(year) that returns True if the given year is a leap year
# and False otherwise, using these rules:
# - The year is a leap year if it is divisible by 4,
#   except for years that are divisible by 100,
#   unless they are also divisible by 400.

# Answer:
def is_leap(year):
    return (year % 4 == 0) and (year % 100 != 0 or year % 400 == 0)

# Example usage (you can change the year value to test):
year = int(input().strip())
print(is_leap(year))
```

```python
# Question 7:
# The program reads an integer n from input.
# Print the numbers from 1 to n (inclusive) together on one line as a single string,
# without any spaces. For example, if n = 3, the output should be:
# 123

# Answer:
if __name__ == '__main__':
    n = int(input().strip())
    result = ''
    for i in range(1, n + 1):
        result += str(i)
    print(result)
```
